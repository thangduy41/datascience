```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

# Processing data

- Đọc file csv
- Chỉ lấy những hàng có Location và Price
- Giữ lại các hàng có ít nhất 1 trong các giá trị Width, Length, Bedrooms, Bathrooms khác null
- Bỏ qua các hàng có Price ảo ( mức ảo tự quy định)
- Loại bỏ hàng trùng

```python
df = pd.read_csv('data.csv')
df = df[(df['District'].notnull())&(df['Province'].notnull())&(df['Price'].notnull())]
df = df[(df['Width'].notnull())|(df['Length'].notnull())|(df['Bedrooms'].notnull())|(df['Bathrooms'].notnull())]
df = df[(df['Price']>=100000000)&(df['Price']<=100000000000)]
df = df.drop_duplicates()
print(len(df))
```

- Tồn tại giá trị âm

Đếm số lượng các hàng có giá trị ảo theo Width, Length, Bedrooms, Bathrooms (mức ảo tự quy định)

```python
df1=df[df['Width']>=100]
df2=df[df['Length']>=100]
df3=df[df['Bedrooms']>=30]
df4=df[df['Bathrooms']>=10]
print(len(df1),len(df2),len(df3),len(df4))
```

```python
df.loc[df['Width'] >100, 'Width'] = df['Width'].median()
df.loc[df['Length'] >100, 'Length'] = df['Length'].median()
df.loc[df['Bedrooms'] >30, 'Bedrooms'] = df['Bedrooms'].median()
df.loc[df['Bathrooms'] >10, 'Bathrooms'] = df['Bathrooms'].median()
```

Định nghĩa các hàm điền các giá trị âm và null:
- **fill_Null_WL**: xử lý cho trường Width, Length. Giá trị null ở col_1 sẽ được thay bằng *median* của các gía trị ở cột tương ứng với từng *Location* của hàng đó (vd: Width null ở hàng có Location là quận Hoàn Kiếm Hà Nội sẽ được thay bằng median Width của các hàng có Location là Hoàn Kiếm Hà Nội)
- **fill_Null_BB**: xử lý cho trường Bedrooms, Bathrooms. Giá trị null ở col_1 sẽ được thay bằng *median* của các gía trị ở cột tương ứng với từng Area của hàng đó (vd: Bedrooms null ở nhà có diện tích 30m2 sẽ đường thay bằng median Bedrooms của các nhà có diện tích 30m2)
- **fill_Null**: xử lý cả 4 trường. Những giá trị nào sau khi xử lý bằng 2 hàm fill_Null_WL và fill_Null_BB ở trên mà vẫn là null thì thay bằng median của tất cả dữ liệu trong trường đó

```python
def fill_Null_WL(df,col_1,col_2,col_3):
    df.loc[df[col_1] <= 0,col_1] = np.nan
    df[col_1] = (df[col_1].fillna(df.groupby([col_2,col_3])[col_1].transform("median")))
def fill_Null_BB(df,col_1,col_2):
    df.loc[df[col_1] <= 0,col_1] = np.nan
    df[col_1] = (df[col_1].fillna(df.groupby(col_2)[col_1].transform("median")))
def fill_Null(df,col_1):
    df[col_1] = (df[col_1].fillna(df[col_1].median()))
```

```python
fill_Null_WL(df,'Width','District','Province')
fill_Null_WL(df,'Length','District','Province')
fill_Null(df,'Width')
fill_Null(df,'Length')
```

Tạo cột Area

```python
df['Area'] = np.round(df['Width']*df['Length'])
print(df.head())
```

```python
fill_Null_BB(df,'Bedrooms','Area')
fill_Null_BB(df,'Bathrooms','Area')

fill_Null(df,'Bedrooms')
fill_Null(df,'Bathrooms')
print(df[-10:])
```
