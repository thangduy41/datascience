```python
# doc tu file du lieu crawl, trich xuat thanh du lieu csv

import findspark
findspark.init()

import pyspark
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
import json

spark = SparkSession\
   .builder\
   .appName("Txt_to_csv")\
   .master("yarn")\
   .config("spark.submit.deployMode","client")\
   .enableHiveSupport()\
   .getOrCreate()

print('register success')
df = spark.read.text("/user/hadoopuser/data_crawl/data_muabannhadat.txt")
print(df.count())


def process_replace(datas):
    out_put = []
    for line in datas:
        data = line[0]
        try:
            data = data.replace('"', '%')
            data = data.replace("'", '"')
            data = data.replace('\\', '')
            data = data.replace('None', '"None"')
            data = json.loads(data)
            
        
            price = data["price"]
            district = data["location"]["parent"]["title"]
            province = data["location"]["parent"]["parent"]["title"]
            width = length = bed = bath = None
            for i in range(len(data["data_properties"])):

                if data["data_properties"][i]["label"]=="Width":
                    if data["data_properties"][i]["value"] != 'None':
                        width = float(data["data_properties"][i]["value"])
                if data["data_properties"][i]["label"]=="Length":
                    if data["data_properties"][i]["value"] != 'None':
                        length = float(data["data_properties"][i]["value"])
                if data["data_properties"][i]["label"]=="Bedrooms":
                    if data["data_properties"][i]["value"] != 'None':
                        bed = int(data["data_properties"][i]["value"])
                if data["data_properties"][i]["label"]=="Bathrooms":
                    if data["data_properties"][i]["value"] != 'None':
                        bath = int(data["data_properties"][i]["value"])
             
            out_put.append((width, length, bed, bath, district, province, price))
              
        except Exception as e:
            continue
  
    return iter(out_put)

schema = StructType([ \
    StructField("Width",DoubleType(),True), \
    StructField("Length",DoubleType(),True), \
    StructField("Bedrooms",IntegerType(),True), \
    StructField("Bathrooms", IntegerType(), True), \
    StructField("District", StringType(), True), \
    StructField("Province", StringType(), True), \
    StructField("Price", LongType(), True)\
  ])


df1 = df.rdd.mapPartitions(process_replace)
df2 = spark.createDataFrame(df1, schema)
print(df2.count())
df2.show()

df2.repartition(4).write.format('csv')\
                  .option('header',True)\
                  .mode('overwrite')\
                  .option('sep',',')\
                  .save('/user/hadoopuser/csv_data')

print('Done!!!!')
spark.stop()
```
