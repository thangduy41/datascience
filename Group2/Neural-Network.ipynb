```python
#import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#import Data
Data = pd.read_csv('data_normalize.csv')
Data.head(5).T
#get some information about our Data-Set
Data.info()
Data.describe().transpose()
```

```python
#visualizing house prices
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,1,1)
sns.distplot(Data['Price'])
fig.add_subplot(2,1,2)
sns.boxplot(Data['Price'])
plt.tight_layout()
#visualizing square footage of (home,lot,above and basement)
fig = plt.figure(figsize=(16,5))
fig.add_subplot(2,3,1)
sns.scatterplot(Data['Width'], Data['Price'])
fig.add_subplot(2,3,2)
sns.scatterplot(Data['Length'],Data['Price'])
fig.add_subplot(2,3,3)
sns.scatterplot(Data['Bedrooms'],Data['Price'])
fig.add_subplot(2,3,4)
sns.scatterplot(Data['Bathrooms'],Data['Price'])
fig.add_subplot(2,3,5)
sns.scatterplot(Data['Distribute'],Data['Price'])
#visualizing bedrooms,bathrooms,floors,grade
fig = plt.figure(figsize=(15,7))
fig.add_subplot(2,3,1)
sns.countplot(Data['Width'])
fig.add_subplot(2,3,2)
sns.countplot(Data['Length'])
fig.add_subplot(2,3,3)
sns.countplot(Data['Bedrooms'])
fig.add_subplot(2,3,4)
sns.countplot(Data['Bathrooms'])
fig.add_subplot(2,3,4)
sns.countplot(Data['Distribute'])
plt.tight_layout()
```

```python
# check if there are any Null values
Data.isnull().sum()
# drop some unnecessary columns
Data = Data.drop('District',axis=1)
Data = Data.drop('Province',axis=1)
```

```python
X = Data.drop('Price',axis =1).values
y = Data['Price'].values/10e9
#splitting Train and Test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
y
```

```python
# Multiple Liner Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
#evaluate the model (intercept and slope)
print(regressor.intercept_)
print(regressor.coef_)
#predicting the test set result
y_pred = regressor.predict(X_test)
#put results as a DataFrame
coeff_df = pd.DataFrame(regressor.coef_, Data.drop('Price',axis =1).columns, columns=['Coefficient']) 
coeff_df
```

```python
# visualizing residuals
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.distplot(residuals)
```

```python
#compare actual output values with predicted values
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
# evaluate the performance of the algorithm (MAE - MSE - RMSE)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

```python
#Model 2: Keras Regressions
# Creating a Neural Network Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
# having 19 neuron is based on the number of available features
model = Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=256,epochs=300)
model.summary()
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8))
```

```python
y_pred = model.predict(X_test)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
# Visualizing Our predictions
fig = plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred)
# Perfect predictions
plt.plot(y_test,y_test,'r')
```
