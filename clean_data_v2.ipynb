```python
import pandas as pd
import numpy as np # Cần cho np.nan

# --- Định nghĩa tên cột ---
# Lưu ý: Tên cột phòng tắm có dấu phẩy và khoảng trắng, cần dùng chính xác
bedroom_col = 'Số phòng ngủ'
bathroom_col = 'Số phòng tắm, vệ sinh' # Đảm bảo tên cột chính xác như trong file CSV của bạn
cols_to_process = [bedroom_col, bathroom_col]

# --- Kiểm tra sự tồn tại của các cột ---
missing_cols = [col for col in cols_to_process if col not in df.columns]
if missing_cols:
    print(f"Lỗi: Không tìm thấy các cột sau trong DataFrame: {missing_cols}")
else:
    print(f"--- Bắt đầu xử lý cột: '{bedroom_col}' và '{bathroom_col}' ---")

    for col in cols_to_process:
        print(f"\n--- Đang xử lý cột: {col} ---")
        original_dtype = df[col].dtype
        print(f"Kiểu dữ liệu gốc: {original_dtype}")

        # --- Bước 1: Loại bỏ chữ ' phòng' và khoảng trắng thừa ---
        # Chỉ thực hiện nếu cột là kiểu object (thường chứa string)
        if pd.api.types.is_object_dtype(df[col]):
            try:
                # Ghi nhận số lượng giá trị là chuỗi rỗng hoặc chỉ chứa khoảng trắng trước khi xử lý
                # để phân biệt với NaN gốc hoặc lỗi chuyển đổi số sau này
                empty_strings_before = df[col].str.strip().eq('').sum()

                # Thay thế ' phòng' bằng rỗng, sau đó xóa khoảng trắng ở đầu/cuối
                df[col] = df[col].str.replace(' phòng', '', regex=False).str.strip()
                print(f"1. Đã loại bỏ ' phòng' và khoảng trắng thừa. Có {empty_strings_before} giá trị là chuỗi rỗng ban đầu.")
            except Exception as e:
                print(f"Lỗi ở Bước 1 (Loại bỏ text): {e}")
                df[col] = None # Đánh dấu lỗi, bỏ qua xử lý cột này
                continue
        elif pd.api.types.is_numeric_dtype(df[col]):
             print("1. Cột đã là kiểu số, bỏ qua bước loại bỏ text.")
        else:
             print("1. Kiểu dữ liệu cột không xác định, cố gắng chuyển sang string.")
             df[col] = df[col].astype(str).str.strip()


        # --- Bước 2: Chuyển đổi sang số, ép lỗi/rỗng thành NaN ---
        if df[col] is not None:
            # Đếm số NaN trước khi chuyển đổi số
            nan_before = df[col].isna().sum()
            # Thực hiện chuyển đổi, chuỗi rỗng ('') và lỗi sẽ thành NaN
            numeric_series = pd.to_numeric(df[col], errors='coerce')
            # Đếm số NaN sau khi chuyển đổi
            nan_after = numeric_series.isna().sum()
            # Số lượng giá trị mới thành NaN = (NaN sau) - (NaN trước)
            # Điều này bao gồm cả chuỗi rỗng và text không hợp lệ
            new_nan_count = nan_after - nan_before

            df[col] = numeric_series # Gán kết quả số (hoặc NaN)

            if new_nan_count > 0:
                 print(f"2. Đã chuyển đổi sang dạng số. {new_nan_count} giá trị rỗng/không phải số đã được chuyển thành NaN.")
            else:
                 print(f"2. Đã chuyển đổi sang dạng số (không có giá trị mới nào thành NaN).")

        # --- Bước 3: Xử lý giá trị không phải số nguyên dương ---
        # Mục tiêu là số nguyên DƯƠNG (>0), nên các giá trị <= 0 hoặc NaN đều cần xử lý
        if df[col] is not None and pd.api.types.is_numeric_dtype(df[col]):
            # Tạo mask cho các giá trị không hợp lệ (<= 0 hoặc NaN)
            # Lưu ý: isna() dùng cho float, isna() hoặc < 1 dùng cho số sau khi coerce
            invalid_mask = (df[col] <= 0) | (df[col].isna())
            invalid_count = invalid_mask.sum()

            if invalid_count > 0:
                # Thay thế các giá trị không hợp lệ bằng pd.NA (giá trị thiếu chuẩn của Pandas cho Int64)
                # Chúng ta dùng np.nan ở đây vì cột vẫn là float, sẽ được chuyển thành pd.NA ở bước sau
                df.loc[invalid_mask, col] = np.nan
                print(f"3. Tìm thấy và thay thế {invalid_count} giá trị không phải nguyên dương (<=0 hoặc NaN) bằng giá trị thiếu (NaN).")
            else:
                print("3. Tất cả giá trị đều là số dương.")
        elif df[col] is None:
            print("3. Bỏ qua xử lý giá trị không dương do lỗi ở bước trước.")
        else:
            print("3. Bỏ qua xử lý giá trị không dương do cột không phải dạng số.")

        # --- Bước 4: Chuyển đổi sang kiểu số nguyên Nullable (Int64) ---
        # Kiểu Int64 có thể chứa cả số nguyên và giá trị thiếu (pd.NA)
        if df[col] is not None and pd.api.types.is_numeric_dtype(df[col]):
             try:
                 df[col] = df[col].astype('Int64')
                 print(f"4. Đã chuyển đổi cột sang kiểu dữ liệu '{df[col].dtype}'.")
             except Exception as e:
                 # Lỗi có thể xảy ra nếu vẫn còn NaN dạng float đặc biệt không tương thích
                 print(f"Lỗi khi chuyển đổi cột sang Int64: {e}. Thử lại bằng cách fillna tạm thời.")
                 # Thử cách khác nếu lỗi: fill NaN bằng 0 tạm thời, chuyển kiểu, rồi đổi 0 lại thành NA nếu cần
                 # Hoặc đơn giản là để lại kiểu float nếu Int64 không thành công
                 # df[col] = df[col].fillna(-999).astype(int).replace(-999, pd.NA) # Cách xử lý phức tạp hơn

        elif df[col] is None:
             print("4. Bỏ qua chuyển đổi Int64 do lỗi ở bước trước.")


    # --- Kiểm tra kết quả tổng thể ---
    print("\n--- Kiểm tra DataFrame sau khi xử lý ---")
    print("5 dòng đầu của DataFrame:")
    print(df.head())
    print("\nThông tin DataFrame:")
    df.info()
    # Kiểm tra số lượng NA trong các cột vừa xử lý
    print("\nSố lượng giá trị NA (thiếu/không hợp lệ) trong các cột đã xử lý:")
    print(df[cols_to_process].isna().sum())
    # print("\nThống kê mô tả (cho các giá trị hợp lệ còn lại):")
    # print(df[cols_to_process].describe())

    # --- (Tùy chọn) Lưu DataFrame đã cập nhật ---
    # output_filename_rooms_int = 'batdongsan_cleaned_rooms_int.csv'
    # try:
    #     df.to_csv(output_filename_rooms_int, index=False, encoding='utf-8-sig')
    #     print(f"\nĐã lưu DataFrame đã xử lý vào file: '{output_filename_rooms_int}'")
    # except Exception as e:
    #     print(f"\nLỗi khi lưu file: {e}")
```

```python
import pandas as pd
import numpy as np

# Tên cột cần xử lý
road_col = 'Đường vào'

# Kiểm tra xem cột có tồn tại không
if road_col not in df.columns:
    print(f"Lỗi: Không tìm thấy cột '{road_col}' trong DataFrame.")
else:
    print(f"--- Bắt đầu xử lý cột: '{road_col}' ---")
    original_dtype = df[road_col].dtype
    print(f"Kiểu dữ liệu gốc: {original_dtype}")

    # Cờ để kiểm tra xem có lỗi xảy ra không
    processing_error = False

    # --- Bước 1: Làm sạch chuỗi (loại bỏ ' m', thay ',', strip) ---
    # Chỉ thực hiện nếu cột là kiểu object hoặc string
    if pd.api.types.is_object_dtype(df[road_col]) or pd.api.types.is_string_dtype(df[road_col]):
        try:
            # Tạo bản sao tạm để xử lý, tránh lỗi SettingWithCopyWarning
            temp_col = df[road_col].astype(str).copy()
            temp_col = temp_col.str.replace(' m', '', regex=False) # Loại bỏ ' m'
            temp_col = temp_col.str.replace(',', '.', regex=False) # Thay thế ',' bằng '.'
            temp_col = temp_col.str.strip() # Xóa khoảng trắng thừa
            df[road_col] = temp_col # Gán lại cột đã xử lý
            print("1. Đã loại bỏ ' m', thay thế ',' bằng '.', và xóa khoảng trắng.")
        except Exception as e:
            print(f"Lỗi ở Bước 1 (Làm sạch chuỗi): {e}")
            processing_error = True
    elif pd.api.types.is_numeric_dtype(df[road_col]):
         print("1. Cột đã là kiểu số, bỏ qua làm sạch chuỗi.")
    else:
         print(f"1. Kiểu dữ liệu cột ({original_dtype}) không xác định, cố gắng chuyển sang string.")
         try:
             df[road_col] = df[road_col].astype(str).str.strip()
         except Exception as e:
             print(f"Lỗi khi ép kiểu sang string: {e}")
             processing_error = True

    # --- Bước 2: Chuyển đổi sang số thực (float), ép lỗi/rỗng thành NaN ---
    if not processing_error:
        try:
            nan_before = df[road_col].replace('', np.nan).isna().sum()
            numeric_series = pd.to_numeric(df[road_col], errors='coerce')
            nan_after = numeric_series.isna().sum()
            # Đếm số giá trị mới thành NaN (rỗng hoặc text lỗi)
            new_nan_count = nan_after - (nan_before if not pd.api.types.is_numeric_dtype(original_dtype) else df[road_col].isna().sum())

            df[road_col] = numeric_series # Gán kết quả số

            if new_nan_count > 0:
                 print(f"2. Đã chuyển đổi sang số thực. {new_nan_count} giá trị rỗng/không phải số đã thành NaN.")
            else:
                 print(f"2. Đã chuyển đổi sang số thực (không có giá trị mới nào thành NaN).")
        except Exception as e:
            print(f"Lỗi ở Bước 2 (Chuyển đổi số): {e}")
            processing_error = True


    # --- Bước 3: Xử lý giá trị không dương (<= 0) ---
    if not processing_error and pd.api.types.is_numeric_dtype(df[road_col]):
        try:
            invalid_mask = (df[road_col] <= 0)
            invalid_count = invalid_mask.sum()
            if invalid_count > 0:
                 df.loc[invalid_mask, road_col] = np.nan # Đặt giá trị không dương thành NaN
                 print(f"3. Tìm thấy và thay thế {invalid_count} giá trị không dương (<=0) bằng NaN.")
            else:
                print("3. Không tìm thấy giá trị không dương.")
        except Exception as e:
             print(f"Lỗi ở Bước 3 (Xử lý giá trị không dương): {e}")
             processing_error = True

    # --- Bước 4: Chuyển đổi sang kiểu số nguyên Nullable (Int64) ---
    # !!! Cảnh báo về việc mất dữ liệu thập phân !!!
    print("\n--- CẢNH BÁO ---")
    print("Bước 4 sẽ chuyển đổi sang số nguyên (Int64).")
    print("Việc này sẽ làm MẤT PHẦN THẬP PHÂN của chiều rộng đường vào (ví dụ: 3.8 sẽ trở thành 3).")
    print("Đối với các số đo như chiều rộng, thường nên giữ kiểu số thực (float).")
    print("-----------------")

    if not processing_error and pd.api.types.is_numeric_dtype(df[road_col]):
        try:
            # Chuyển đổi sang Int64, phần thập phân sẽ bị cắt bỏ
            df[road_col] = df[road_col].astype('Int64')
            print(f"4. Đã chuyển đổi cột sang kiểu dữ liệu '{df[road_col].dtype}'. Phần thập phân (nếu có) đã bị loại bỏ.")
        except Exception as e:
            print(f"Lỗi khi chuyển đổi cột sang Int64: {e}")
            print("   Giữ nguyên kiểu dữ liệu float do lỗi chuyển đổi Int64.") # Giữ lại float nếu Int64 lỗi


    # --- Kiểm tra kết quả ---
    print("\n--- Kiểm tra DataFrame sau khi xử lý ---")
    if road_col in df.columns and not processing_error:
        print("5 dòng đầu của DataFrame:")
        print(df.head())
        print("\nThông tin DataFrame:")
        df.info()
        print(f"\nSố lượng giá trị NA trong cột '{road_col}': {df[road_col].isna().sum()}")
        # print(f"\nThống kê mô tả cho cột '{road_col}':") # Bỏ comment nếu muốn xem describe
        # print(df[road_col].describe())
    elif road_col not in df.columns:
        print(f"Cột '{road_col}' không tồn tại để hiển thị kết quả.")
    else:
         print("Không thể hiển thị kết quả do có lỗi trong quá trình xử lý.")

    # --- (Tùy chọn) Lưu DataFrame đã cập nhật ---
    # output_filename_road = 'batdongsan_cleaned_road.csv'
    # if not processing_error:
    #      try:
    #           df.to_csv(output_filename_road, index=False, encoding='utf-8-sig')
    #           print(f"\nĐã lưu DataFrame đã xử lý vào file: '{output_filename_road}'")
    #      except Exception as e:
    #           print(f"\nLỗi khi lưu file: {e}")
```

```python
import pandas as pd
import numpy as np

# --- Định nghĩa tên cột ---
bedroom_col = 'Số phòng ngủ'
bathroom_col = 'Số phòng tắm, vệ sinh'
cols_to_impute = [bedroom_col, bathroom_col]

print("--- Bắt đầu điền giá trị thiếu bằng Trung vị (Median) ---")

# --- Kiểm tra sự tồn tại và kiểu dữ liệu của các cột ---
proceed = True
for col in cols_to_impute:
    if col not in df.columns:
        print(f"Lỗi: Không tìm thấy cột '{col}'. Bỏ qua xử lý.")
        proceed = False
        break
    # Kiểm tra xem cột đã là kiểu số chưa (cần thiết để tính median)
    if not pd.api.types.is_numeric_dtype(df[col]):
        print(f"Cảnh báo: Cột '{col}' hiện không phải là kiểu số (kiểu {df[col].dtype}).")
        print("   Vui lòng đảm bảo cột này đã được chuyển đổi sang dạng số (ví dụ: Int64, float) trước khi chạy bước này.")
        proceed = False
        break

if proceed:
    for col in cols_to_impute:
        print(f"\n--- Đang xử lý cột: {col} ---")

        # --- Bước 1: Kiểm tra xem có giá trị thiếu không ---
        # Sử dụng isna() cho cả NaN (float) và NA (Int64)
        na_count = df[col].isna().sum()
        print(f"1. Số lượng giá trị thiếu (NA/NaN) hiện tại: {na_count}")

        if na_count > 0:
            # --- Bước 2: Tính toán giá trị Trung vị ---
            # .median() sẽ tự động bỏ qua các giá trị NA/NaN khi tính toán
            median_value = df[col].median()

            # Xử lý trường hợp cột toàn NA (median_value sẽ là NaN)
            if pd.isna(median_value):
                fill_value = 0 # Hoặc một giá trị mặc định khác bạn cho là hợp lý
                print(f"   Cảnh báo: Không thể tính trung vị cho cột '{col}' (có thể tất cả đều là NA). Sử dụng giá trị mặc định là {fill_value}.")
            else:
                # Làm tròn và chuyển thành số nguyên để phù hợp với số phòng
                fill_value = int(round(median_value))
                print(f"2. Giá trị Trung vị tính được (làm tròn): {fill_value}")

            # --- Bước 3: Điền giá trị thiếu bằng Trung vị đã tính ---
            df[col].fillna(fill_value, inplace=True)
            print(f"3. Đã điền {na_count} giá trị NA/NaN bằng Trung vị ({fill_value}).")

            # --- Bước 4: Đảm bảo cột là kiểu Int64 ---
            # Sau khi fillna, nếu cột ban đầu là Int64 và không còn NA, nó có thể thành int64
            # Hoặc nếu cột là float và fill bằng int, nó vẫn là float. Cần ép lại Int64.
            try:
                df[col] = df[col].astype('Int64')
                print(f"4. Đã xác nhận/chuyển đổi lại cột sang kiểu '{df[col].dtype}'.")
            except Exception as e:
                print(f"4. Lỗi khi chuyển đổi lại sang Int64 sau fillna: {e}")
        else:
            print("2. Không có giá trị thiếu trong cột này để điền.")

    # --- Kiểm tra kết quả tổng thể ---
    print("\n--- Kiểm tra DataFrame sau khi điền giá trị thiếu ---")
    print("5 dòng đầu của DataFrame:")
    print(df.head())
    print("\nThông tin DataFrame:")
    df.info()
    print("\nSố lượng giá trị NA còn lại trong các cột đã xử lý:")
    # Kết quả mong đợi là 0 cho các cột đã xử lý thành công
    print(df[cols_to_impute].isna().sum())
    # print("\nThống kê mô tả:") # Bỏ comment nếu muốn xem
    # print(df[cols_to_impute].describe())

    # --- (Tùy chọn) Lưu DataFrame đã cập nhật ---
    # output_filename_imputed = 'batdongsan_imputed.csv'
    # try:
    #     df.to_csv(output_filename_imputed, index=False, encoding='utf-8-sig')
    #     print(f"\nĐã lưu DataFrame đã xử lý vào file: '{output_filename_imputed}'")
    # except Exception as e:
    #     print(f"\nLỗi khi lưu file: {e}")

else:
    print("\nKhông thể thực hiện điền giá trị do thiếu cột hoặc cột chưa được chuyển đổi sang dạng số.")
```
