```python
import numpy as np
import pandas as pd # Đảm bảo pandas đã được import

# --- Giả sử DataFrame 'df' đã tồn tại sau khi đọc file CSV gốc ---

# --- Bước 1: Xóa bỏ các hàng có "Giá thỏa thuận" ---
gia_column_name = 'Giá'

if gia_column_name not in df.columns:
    print(f"Lỗi: Không tìm thấy cột '{gia_column_name}' trong DataFrame.")
else:
    rows_before = len(df)
    mask = df[gia_column_name] != 'Giá thỏa thuận'
    df = df[mask].copy()
    rows_after = len(df)
    print(f"Đã xóa {rows_before - rows_after} hàng có giá trị 'Giá thỏa thuận'.")

    # --- Bước 2: Thay thế dấu phẩy (,) bằng dấu chấm (.) ---
    if pd.api.types.is_string_dtype(df[gia_column_name]):
         df[gia_column_name] = df[gia_column_name].str.replace(',', '.', regex=False)
         print("Đã thay thế dấu ',' bằng '.' trong cột Giá.")
    else:
         try:
             df[gia_column_name] = df[gia_column_name].astype(str).str.replace(',', '.', regex=False)
             print("Đã ép kiểu sang string và thay thế dấu ',' bằng '.' trong cột Giá.")
         except Exception as e:
             print(f"Không thể thay thế dấu ',' do lỗi kiểu dữ liệu: {e}")

    # --- Bước 3: Chuyển đổi đơn vị và loại bỏ text ---
    def convert_price_unit(price_str):
        if pd.isna(price_str) or isinstance(price_str, (int, float)):
            return price_str
        if not isinstance(price_str, str):
             return np.nan
        price_str = price_str.strip().lower()
        try:
            if 'tỷ' in price_str:
                num_part = price_str.replace('tỷ', '').strip()
                return float(num_part) * 1e9
            elif 'triệu' in price_str:
                num_part = price_str.replace('triệu', '').strip()
                return float(num_part) * 1e6
            else:
                return float(price_str)
        except ValueError:
            return np.nan

    df[gia_column_name] = df[gia_column_name].apply(convert_price_unit)
    print("Đã áp dụng hàm chuyển đổi đơn vị (tỷ/triệu) và loại bỏ text.")

    # --- Bước 4: Chuyển đổi kiểu dữ liệu sang số thực ---
    df[gia_column_name] = pd.to_numeric(df[gia_column_name], errors='coerce')
    print(f"Đã chuyển đổi cột '{gia_column_name}' sang kiểu số thực.")

    # --- Kiểm tra kết quả ---
    print("\n5 dòng đầu của DataFrame sau khi xử lý cột 'Giá':")
    print(df.head())
    print("\nThông tin kiểu dữ liệu của DataFrame sau khi xử lý cột 'Giá':")
    df.info()
    # print("\nThống kê mô tả cho cột 'Giá' sau khi xử lý:") # Tạm ẩn đi cho đỡ dài
    # print(df[gia_column_name].describe())

```

```python
import pandas as pd
import numpy as np

# Tên cột cần xử lý
area_column_name = 'Diện tích'

# Kiểm tra xem cột có tồn tại trong DataFrame không
if area_column_name not in df.columns:
    print(f"Lỗi: Không tìm thấy cột '{area_column_name}' trong DataFrame.")
else:
    print(f"--- Bắt đầu xử lý lại cột: '{area_column_name}' ---")

    # --- Bước 1: Loại bỏ đơn vị " m²" ---
    # Đảm bảo cột là kiểu string và thay thế ' m²' bằng chuỗi rỗng ''
    try:
        df[area_column_name] = df[area_column_name].astype(str).str.replace(' m²', '', regex=False)
        print("1. Đã loại bỏ đơn vị ' m²'.")
    except Exception as e:
        print(f"Lỗi khi loại bỏ đơn vị: {e}")
        # Dừng xử lý nếu có lỗi nghiêm trọng ở bước này
        df[area_column_name] = None # Đặt cột thành None hoặc xử lý khác nếu cần


    # --- Bước 2: Chuẩn hóa định dạng số (Xử lý dấu chấm và phẩy) ---
    # Quan trọng: Thực hiện theo đúng thứ tự giả định:
    # 1. Xóa dấu chấm (.) được coi là dấu phân cách hàng nghìn.
    # 2. Thay thế dấu phẩy (,) được coi là dấu thập phân bằng dấu chấm (.).
    if df[area_column_name] is not None: # Chỉ chạy nếu bước 1 thành công
        try:
            # 2a. Xóa dấu chấm (.) phân cách hàng nghìn
            df[area_column_name] = df[area_column_name].str.replace('.', '', regex=False)
            print("2a. Đã loại bỏ dấu '.' (coi là phân cách hàng nghìn).")

            # 2b. Thay thế dấu phẩy (,) thập phân bằng dấu chấm (.)
            df[area_column_name] = df[area_column_name].str.replace(',', '.', regex=False)
            print("2b. Đã thay thế dấu ',' (coi là thập phân) bằng dấu '.'.")
        except Exception as e:
            print(f"Lỗi khi chuẩn hóa định dạng số: {e}")
            df[area_column_name] = None # Đặt cột thành None hoặc xử lý khác nếu cần


    # --- Bước 3: Chuyển đổi kiểu dữ liệu sang số thực (float) ---
    if df[area_column_name] is not None: # Chỉ chạy nếu bước 2 thành công
        original_type = df[area_column_name].dtype
        # errors='coerce' sẽ biến các giá trị không thể chuyển đổi thành NaN
        df[area_column_name] = pd.to_numeric(df[area_column_name], errors='coerce')
        print(f"3. Đã chuyển đổi cột '{area_column_name}' sang kiểu '{df[area_column_name].dtype}'.")


    # --- Bước 4: Kiểm tra sau chuyển đổi ---
    print("\n--- Kiểm tra kết quả sau khi xử lý ---")
    if area_column_name in df.columns and df[area_column_name] is not None:
        # Hiển thị 5 dòng đầu để xem kết quả trực quan
        print("5 dòng đầu của DataFrame:")
        print(df.head())

        # Kiểm tra thông tin tổng quan
        print("\nThông tin DataFrame:")
        df.info()

        # Kiểm tra xem có giá trị NaN nào được tạo ra không
        nan_count = df[area_column_name].isnull().sum()
        if nan_count > 0:
            print(f"\nCảnh báo: Có {nan_count} giá trị không hợp lệ (NaN) trong cột '{area_column_name}' sau khi chuyển đổi.")

        # Hiển thị thống kê mô tả
        print(f"\nThống kê mô tả cho cột '{area_column_name}':")
        print(df[area_column_name].describe())

        # Kiểm tra cụ thể các giá trị 0 hoặc âm
        zero_area_count = (df[area_column_name] == 0).sum()
        negative_area_count = (df[area_column_name] < 0).sum()
        if zero_area_count > 0:
            print(f"- Số lượng bản ghi có diện tích bằng 0: {zero_area_count}")
        if negative_area_count > 0:
            print(f"- Số lượng bản ghi có diện tích âm: {negative_area_count}")
    else:
        print("Không thể hiển thị kết quả do có lỗi ở các bước xử lý trước.")

```

```python
import pandas as pd

# Tên cột gốc và các cột mới
location_col = 'Địa điểm'
district_col = 'Huyện' # Tên cột mới cho Quận/Huyện
province_col = 'Tỉnh'  # Tên cột mới cho Tỉnh/Thành phố

# Kiểm tra xem cột gốc có tồn tại không
if location_col not in df.columns:
    print(f"Lỗi: Không tìm thấy cột '{location_col}' để xử lý.")
else:
    print(f"--- Bắt đầu xử lý cột '{location_col}' ---")

    try:
        # --- Bước 1: Tách cột 'Địa điểm' thành 2 phần ---
        # Sử dụng .str.split() với delimiter là dấu phẩy ','
        # n=1: Chỉ tách ở dấu phẩy đầu tiên tìm thấy
        # expand=True: Trả về kết quả dưới dạng các cột DataFrame riêng biệt
        split_data = df[location_col].astype(str).str.split(',', n=1, expand=True)

        # --- Bước 2: Tạo các cột mới 'Huyện' và 'Tỉnh' ---
        # Cột đầu tiên (index 0) của kết quả tách là Huyện
        df[district_col] = split_data[0]
        # Cột thứ hai (index 1) của kết quả tách là Tỉnh
        # Sẽ tự động là None nếu không có dấu phẩy để tách
        df[province_col] = split_data[1]
        print(f"1. Đã tách dữ liệu thành cột '{district_col}' và '{province_col}'.")

        # --- Bước 3: Làm sạch khoảng trắng thừa ở cột mới ---
        df[district_col] = df[district_col].str.strip()
        df[province_col] = df[province_col].str.strip()
        print("2. Đã loại bỏ khoảng trắng thừa (nếu có) ở cột 'Huyện' và 'Tỉnh'.")

        # --- Bước 4: Xóa cột 'Địa điểm' gốc ---
        # axis=1 chỉ định xóa cột, inplace=True thực hiện xóa trực tiếp
        df.drop(columns=[location_col], axis=1, inplace=True)
        print(f"3. Đã xóa cột gốc '{location_col}'.")

        # --- Kiểm tra kết quả ---
        print("\n--- Kiểm tra kết quả sau khi xử lý ---")
        print("5 dòng đầu của DataFrame với các cột địa lý mới:")
        print(df.head())
        print("\nThông tin DataFrame cập nhật:")
        df.info()

        # Kiểm tra xem có hàng nào thiếu thông tin Tỉnh không (do không có dấu phẩy)
        missing_province_count = df[province_col].isnull().sum()
        if missing_province_count > 0:
            print(f"\nLưu ý: Có {missing_province_count} hàng bị thiếu thông tin '{province_col}' (có thể do dữ liệu gốc không chứa dấu phẩy).")

    except Exception as e:
        print(f"Đã xảy ra lỗi trong quá trình xử lý: {e}")
        # Có thể thêm xử lý lỗi khác ở đây nếu cần
```

```python
import pandas as pd

print("--- Bắt đầu xử lý dữ liệu trùng lặp ---")

# --- Bước 1: Kiểm tra và đếm số lượng hàng trùng lặp ---
# df.duplicated() trả về một chuỗi Boolean, True cho mỗi hàng là bản sao (trừ lần xuất hiện đầu tiên)
# .sum() sẽ đếm số lượng giá trị True (tức là số hàng trùng lặp)
num_duplicates_before = df.duplicated().sum()

if num_duplicates_before > 0:
    print(f"1. Phát hiện: Tìm thấy {num_duplicates_before} hàng dữ liệu bị trùng lặp hoàn toàn.")

    # --- Bước 2: Loại bỏ các hàng trùng lặp ---
    # Lưu lại số hàng trước khi xóa để thông báo
    rows_before_drop = len(df)

    # df.drop_duplicates() sẽ xóa các hàng trùng lặp
    # Mặc định keep='first', giữ lại hàng đầu tiên và xóa các hàng giống hệt phía sau
    # Gán kết quả lại cho df (thay vì dùng inplace=True) thường được khuyến nghị
    df = df.drop_duplicates()

    # Lấy số hàng sau khi xóa
    rows_after_drop = len(df)

    print(f"2. Thực hiện: Đã xóa {rows_before_drop - rows_after_drop} hàng trùng lặp.")
    print(f"   - Số hàng còn lại: {rows_after_drop}")

    # Kiểm tra lại để chắc chắn không còn hàng trùng lặp
    num_duplicates_after = df.duplicated().sum()
    if num_duplicates_after == 0:
        print("3. Xác nhận: Không còn hàng nào bị trùng lặp hoàn toàn trong DataFrame.")
    else:
        # Trường hợp này hiếm khi xảy ra nếu drop_duplicates hoạt động đúng
        print(f"Cảnh báo: Vẫn còn {num_duplicates_after} hàng trùng lặp sau khi xóa?")

else:
    print("1. Phát hiện: Không tìm thấy hàng nào bị trùng lặp hoàn toàn.")
    print("2. Thực hiện: Không cần xóa hàng nào.")

# --- Hiển thị thông tin về DataFrame sau khi xử lý ---
print("\n--- Thông tin DataFrame sau khi xử lý trùng lặp ---")
print(f"Kích thước cuối cùng của DataFrame (số hàng, số cột): {df.shape}")
# Bạn có thể bỏ comment dòng dưới nếu muốn xem chi tiết hơn
# df.info()
```
